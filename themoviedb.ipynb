```python
import requests
import pandas as pd
from config import key
```

```python
themoviedb_ids=[]
conn_timeout = 5
read_timeout = 9
timeouts = (conn_timeout, read_timeout)
#run total 1.9mils ids max.
for j in range (0,20):
    #initiate empty df
    df = pd.DataFrame({'themoviedb_id':[],
                   'imdb_id':[],
                   'belongs_to_collection':[],
                   'budget':[],
                   'popularity':[],
                   'original_language':[],
                   'production_companies_name':[],
                   'production_companies_id':[],
                   'production_countries':[],
                   'revenue':[],
                   'status':[],
                   'vote_average':[],
                   'vote_count':[],
                   'overview':[]
                   })
    #run each file 100,000 ids max.
    for i in range(100000*j,100000*(j+1)):
        #data url
        url = f'https://api.themoviedb.org/3/movie/{i}?api_key={key}'
        try:
            data = requests.get(url, timeout=timeouts)
        except:
            print(f'exception {j}, {i}')
            continue
        #check if page exist
        if data.status_code == 404:
            continue
        else:
            try:
                #note the movies ids
                themoviedb_ids.append(i)
                #convert data to json format
                data = data.json()
                #check data conditions
                if (len(data['production_companies'])==0):
                    production_company_name = None
                    production_company_id = None
                else:
                    production_company_name = data['production_companies'][0]['name']
                    production_company_id = data['production_companies'][0]['id']
                if (len(data['production_countries'])==0):
                    production_countries = None
                else:
                    production_countries = data['production_countries'][0]['iso_3166_1']
                if (data['belongs_to_collection']==None):
                    belongs_to_collection = None
                else:
                    belongs_to_collection =data['belongs_to_collection']['name']
                #concat new data to dataframe
                df = pd.concat([df ,
                                pd.DataFrame([[i,data['imdb_id'],belongs_to_collection,data['budget'],data['popularity'],data['original_language'],
                                                    production_company_name,production_company_id,
                                                    production_countries,data['revenue'],
                                                    data['status'],data['vote_average'],data['vote_count'],data['overview']   ]], 
                                        columns=['themoviedb_id','imdb_id','belongs_to_collection','budget','popularity','original_language',
                                                'production_companies_name','production_companies_id','production_countries','revenue',
                                                'status','vote_average','vote_count','overview']) 
                                ],
                ignore_index=True)
                print (f'success {j}, {i}')
            except:
                pass
    #export partial file to excel        
    df.to_csv(f'./data/themoviedb_id_{100000*j}_to_{100000*(j+1)}.csv',index=False)

print(len(themoviedb_ids))
```
