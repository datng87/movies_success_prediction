```python
import requests
import pandas as pd

df2 = pd.read_csv('./data/subset_movie_data_new.csv')
df2 =df2[['tconst','startYear']]
unique_ids = df2.tconst.unique()
unique_ids = unique_ids[0:42103]

df1 = pd.DataFrame({'tconst':[],
                       'Rated':[],
                   'Plot':[],
                   'Awards':[],
                   'Metascore':[],
                   'DVD':[],
                   'BoxOffice':[],
                   'Actors':[],
                   'Released':[],
                   'Director':[],
                   'Writer':[],
                   'rating1source':[],'rating1value':[],'rating2source':[],'rating2value':[],'rating3source':[],'rating3value':[]
                   })
conn_timeout = 4
read_timeout = 7
timeouts = (conn_timeout, read_timeout)

count=0
for id in unique_ids:
    count+=1
    
    try:
        url1 = f'http://www.omdbapi.com/?i={id}&apikey=57c63d19'
        omdbdata = requests.get(url1,timeout=timeouts)
        print (f'success {id}, movie#{count}')
    except:
        print(f'exception {id}')
        continue
    if omdbdata.status_code == 404:
        continue
    else:
        try:
            omdbdata = omdbdata.json()    
            if len(omdbdata['Ratings'])==0:
                rating1source = 0
                rating1value = 0
                rating2source = 0
                rating2value = 0
                rating3source = 0
                rating3value = 0
            if len(omdbdata['Ratings'])==1:
                rating1source = omdbdata['Ratings'][0]['Source']
                rating1value = omdbdata['Ratings'][0]['Value']
                rating2source = 0
                rating2value = 0
                rating3source = 0
                rating3value = 0
            if len(omdbdata['Ratings'])==2:
                rating1source = omdbdata['Ratings'][0]['Source']
                rating1value = omdbdata['Ratings'][0]['Value']
                rating2source = omdbdata['Ratings'][1]['Source']
                rating2value = omdbdata['Ratings'][1]['Value']
                rating3source = 0
                rating3value = 0
            if len(omdbdata['Ratings'])>=3:
                rating1source = omdbdata['Ratings'][0]['Source']
                rating1value = omdbdata['Ratings'][0]['Value']
                rating2source = omdbdata['Ratings'][1]['Source']
                rating2value = omdbdata['Ratings'][1]['Value']
                rating3source = omdbdata['Ratings'][2]['Source']
                rating3value = omdbdata['Ratings'][2]['Value']
        except:
            rating1source = 0
            rating1value = 0
            rating2source = 0
            rating2value = 0
            rating3source = 0
            rating3value = 0
        try:
            dvd = omdbdata['DVD']
        except:
            dvd = 0
        try:
            BoxOffice = omdbdata['BoxOffice']
        except:
            BoxOffice = 0    
        try:
            Rated = omdbdata['Rated']
        except:
            Rated = 0  
        try:
            plot = omdbdata['Plot']
        except:
            plot = 0  
        try:
            Awards = omdbdata['Awards']
        except:
            Awards = 0  
        try:
            Metascore = omdbdata['Metascore']
        except:
            Metascore = 0  
        try:
            Actors=omdbdata['Actors']
        except:
            Actors = 0  
        try:
            Released = omdbdata['Released']
        except:
            Released = 0  
        try:
            Director = omdbdata['Director']
        except:
            Director = 0  
        try:
            Writer = omdbdata['Writer']   
        except:
            Writer = 0  

        df1 = pd.concat([df1 ,
                            pd.DataFrame([[id,Rated,plot,Awards,Metascore,dvd,
                                        BoxOffice,Actors,Released,Director,Writer,
                                        rating1source,rating1value,rating2source,rating2value,rating3source,rating3value]], 
                                                columns=['tconst','Rated','Plot','Awards','Metascore','DVD',
                                                        'BoxOffice','Actors','Released','Director','Writer',
                                                        'rating1source','rating1value','rating2source','rating2value','rating3source','rating3value']) 
                                    ],
                    ignore_index=True)

df1.to_csv('./omdb/omdbdata_0_42103.csv',index=False)    
```
